#### This extracts the month, month name, and year and creates a new column each.

### Q1: Which product category generates the most total sales?

```python
product_sales = df.groupby('Category')['Sales'].sum().sort_values(ascending = False)
product_sales 
```

```python
#chart
plt.figure(figsize=(8,6))
sns.barplot(x =product_sales.index,y=product_sales.values, palette='tab10')
plt.title('Total Sales by Product Category')
plt.xlabel('Category')
plt.ylabel('Total Sales ($)')
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()
```

 **Finding:** Technology is the highest revenue-generating category 
with total sales of $836,154, followed by Furniture($741,999) 
and Office Supplies ($719,047). This shows that despite being 
a retail store, technology products drive the most revenue.

## Q3: Which region is the most profitable?

```python
profit_by_region = df.groupby('Region')['Profit'].sum().sort_values(ascending = False)
profit_by_region
```

```python
sales_by_region = df.groupby('Region')['Sales'].sum()
sales_by_region
```

```python
profit_margin_by_region = (profit_by_region / sales_by_region * 100).round(1)
profit_margin_by_region
```

### Key Insights:
- **West** leads in both sales and profit, with the highest profit margin (14.9%)
- **East** has the second-highest sales and maintains a healthy margin (13.5%)
- **Central** has the third-highest sales but the lowest profit margin (7.9%),
  suggesting high costs or heavy discounting
- **South** has the lowest sales but a better margin than Central (11.9%)

### Conclusion:
Higher sales do not always mean higher profit. The Central region is a clear example 
where strong sales volume does not translate to strong profitability.

### Q5: Are there products that sell a lot but generate little profit?

```python
product_profit = df.groupby('Category')[['Sales','Profit']].sum().sort_values('Profit',ascending = False)
product_profit
product_profit['Profit_Margin %'] = ((product_profit['Profit']/product_profit['Sales'])*100).sort_values().round(2)
product_profit
```

```python
#chart
plt.figure(figsize=(7,6))
sns.barplot(data = product_profit, x = 'Category', y = "Profit_Margin %", palette = 'tab10')
plt.title('Profit Margin by Category')
plt.xlabel('Category')
plt.ylabel('Profit Margin %')
plt.show()
```

```python
problem_products = df[df['Profit']<0].groupby(['Sub-Category','Category'])[['Sales','Profit']].sum()
problem_products.sort_values('Sales')
problem_products['Profit Margin %'] = ((problem_products['Profit']/problem_products['Sales'])*100).sort_values().round(0)
problem_products.sort_values('Profit Margin %')
```

```python
plt.figure(figsize=(14,6))
sns.barplot(data=problem_products,x='Profit Margin %',y='Sub-Category', 
            hue = 'Category',palette ='tab10')
plt.title('Profit Margin % by for Loss-Making Products')
plt.axvline(x=0,color = 'black',linestyle='--')
plt.xlabel('Profit Margin %')
plt.ylabel('Sub-Category')
plt.tight_layout()
plt.show()
```

#### **Finding:** At category level, Furniture has the lowest profit  margin at 2.49% while Technology and Office Supplies stay above 16%. However at sub-category level, Appliances (-255%) and Binders (-107%) are the worst performing products, both under Office Supplies. This reveals that even a healthy category can hide severely loss-making products underneath.

```python


# Group by Year and State
state_year_sales = (df.groupby(['Year', 'State'])['Sales']
                    .sum()
                    .reset_index()
                    .sort_values(['Year', 'Sales'], ascending=[True, False]))

# Get top 5 states overall to track across years
top_states = (df.groupby('State')['Sales']
              .sum()
              .sort_values(ascending=False)
              .head(5)
              .index.tolist())

# Filter only top 5 states
filtered = state_year_sales[state_year_sales['State'].isin(top_states)]

# Plot
fig, ax = plt.subplots(figsize=(12, 6))

for state in top_states:
    state_data = filtered[filtered['State'] == state]
    ax.plot(state_data['Year'], state_data['Sales'], marker='o', label=state, linewidth=2)

ax.set_title('Top 5 States Sales Trend (2014–2017)', fontsize=14)
ax.set_xlabel('Year')
ax.set_ylabel('Total Sales (USD)')
ax.legend(title='State')
ax.set_xticks([2014, 2015, 2016, 2017])
plt.tight_layout()
plt.show()
```

## Sales Trend of Top 5 States (2014–2017)

### Key Observations:
- **California** dominates all other states across all four years, with a strong
  upward trajectory — growing from ~$92,000 in 2014 to ~$146,000 in 2017,
  despite a slight dip in 2015
- **New York** holds a steady second position, dipping in 2016 before recovering
  strongly to ~$94,000 in 2017
- **Texas** started strong at ~$51,000 in 2014 but dropped sharply in 2015,
  recovering gradually to ~$43,000 by 2017
- **Washington** had the most volatile trend — dropping to its lowest point in 2016
  (~$20,000) before surging to ~$65,000 in 2017, overtaking Texas and Pennsylvania
- **Pennsylvania** showed the most stable and gradual growth, rising steadily
  from ~$21,000 in 2014 to ~$43,000 in 2017

### Takeaway:
> California is the clear sales leader and shows consistent long-term growth.
> Washington's sharp 2017 surge is worth investigating — it could reflect
> a new market opportunity or seasonal spike. Texas, despite its size,
> shows inconsistent performance and may need a targeted sales strategy.
