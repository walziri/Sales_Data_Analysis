```python
df['Month'] = df['Order Date'].dt.month
df['Month Name'] = df['Order Date'].dt.strftime('%B')
df['Year'] = df['Order Date'].dt.year
```

### Q1: Which product category generates the most total sales?

```python
product_sales = df.groupby('Category')['Sales'].sum().sort_values(ascending = False)
product_sales 
```

```python
#chart
plt.figure(figsize=(8,6))
sns.barplot(x =product_sales.index,y=product_sales.values, palette='tab10')
plt.title('Total Sales by Product Category')
plt.xlabel('Category')
plt.ylabel('Total Sales ($)')
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()
```

 **Finding:** Technology is the highest revenue-generating category 
with total sales of $836,154, followed by Furniture($741,999) 
and Office Supplies ($719,047). This shows that despite being 
a retail store, technology products drive the most revenue.

## Q2:Which month has the highest sales across all years?

```python
monthly_sales = df.groupby('Month Name')['Sales'].sum().sort_values(ascending = False)
monthly_sales
month_order1 = ['January','February','March','April','May','June','July','August','September','October','November','December']
monthly_sales
monthly_sales = monthly_sales.reindex(month_order1)
monthly_sales
```

```python
#chart
plt.figure(figsize=(15,7))
sns.lineplot(x=monthly_sales.index,y=monthly_sales.values,palette= 'Blues_d',marker = 's')
plt.title('Total Sales by Month')
plt.xlabel('Month')
plt.ylabel('Total Sales($)')
plt.xticks(rotation =30)
plt.tight_layout()
plt.show()
```

**Finding:** November records the highest sales across all years,
which aligns with the Black Friday and holiday shopping season
in the US. Sales are generally low at the start of the year 
and pick up significantly towards Q4.

## Q3: Which region is the most profitable?

```python
profit_by_region = df.groupby('Region')['Profit'].sum().sort_values(ascending = False)
profit_by_region
```

```python
#chart
plt.figure(figsize=(8,6))
sns.barplot(x=profit_by_region.index, y =profit_by_region.values, palette = 'tab10')
plt.title('Profit my Region')
plt.xlabel('Region')
plt.ylabel('Profit')
plt.tight_layout()
plt.show()
```

**Finding:** The West region is the most profitable, which is 
consistent with Q7 where California — located in the West —

generally drives higher profit.

## Q4:Which sub-category receives the highest average discount?

```python
discount = df.groupby('Sub-Category')['Discount'].mean().sort_values(ascending = False)
discount
```

```python
#chart
plt.figure(figsize=(18,7))
sns.barplot(x=discount.index, y = discount.values, palette = 'Blues_d')
plt.title('Average Discount by Sub-Category')
plt.xlabel('Sub-Category')
#plt.xticks(rotation=25)
plt.ylabel('Average Discount')
plt.tight_layout()
plt.show()
```

**Finding:** Binders receive the highest average discount among

### Q5: Are there products that sell a lot but generate little profit?

```python
product_profit = df.groupby('Category')[['Sales','Profit']].sum().sort_values('Profit',ascending = False)
product_profit
product_profit['Profit_Margin %'] = ((product_profit['Profit']/product_profit['Sales'])*100).sort_values().round(2)
product_profit
```

```python
#chart
plt.figure(figsize=(7,6))
sns.barplot(data = product_profit, x = 'Category', y = "Profit_Margin %", palette = 'tab10')
plt.title('Profit Margin by Category')
plt.xlabel('Category')
plt.ylabel('Profit Margin %')
plt.show()
```

```python
problem_products = df[df['Profit']<0].groupby(['Sub-Category','Category'])[['Sales','Profit']].sum()
problem_products.sort_values('Sales')
problem_products['Profit Margin %'] = ((problem_products['Profit']/problem_products['Sales'])*100).sort_values().round(0)
problem_products.sort_values('Profit Margin %')
```

```python
plt.figure(figsize=(14,6))
sns.barplot(data=problem_products,x='Profit Margin %',y='Sub-Category', 
            hue = 'Category',palette ='tab10')
plt.title('Profit Margin % by for Loss-Making Products')
plt.axvline(x=0,color = 'black',linestyle='--')
plt.xlabel('Profit Margin %')
plt.ylabel('Sub-Category')
plt.tight_layout()
plt.show()
```

**Finding:** At category level, Furniture has the lowest profit 
margin at 2.49% while Technology and Office Supplies stay above 
16%. However at sub-category level, Appliances (-255%) and 
Binders (-107%) are the worst performing products, both under 
Office Supplies. This reveals that even a healthy category can 
hide severely loss-making products underneath.

## Q6:How has total sales changed year over year?

```python
yearly_sales = df.groupby('Year')['Sales'].sum()
yearly_sales
```

```python
#Chart
plt.figure(figsize=(8,6))
sns.barplot(x=yearly_sales.index,y=yearly_sales.values,palette='tab10')
plt.title('Total Sales (2014-2017)')
plt.xlabel('Year')
plt.ylabel('Totals Sales')
plt.tight_layout()
plt.show()
```

```python
monthly_trendline = df.groupby(['Year','Month'])['Sales'].sum().reset_index()
monthly_trendline
#chart
plt.figure(figsize=(12,6))
sns.lineplot(data = monthly_trendline,x='Month',y = 'Sales',hue='Year',marker='s',palette='tab10')
plt.xticks(ticks= range(1,13),
           labels = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec'],rotation=25,)
plt.title('Monthly Sales Trend by Year')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.legend(title='Year')
plt.tight_layout()
plt.show()
```

**Finding:** Sales have grown consistently from 2014 to 2017 
with a clear seasonal pattern — sales spike at the end of each 
year (November/December) and dip at the beginning. The overall 
upward trend indicates a healthy, growing business.

## Q7: Which state has the highest number of orders?

```python
order_no= df['State'].value_counts().head(10)
order_no
plt.figure(figsize=(20,8))
sns.barplot(x = order_no.index, y = order_no.values,palette = 'Blues_d')
plt.title('Number or Order by State')
plt.xlabel('State')
plt.ylabel('No. of Orders')
plt.tight_layout()
plt.show()
```

**Finding:** California has the highest number of orders among 
all US states, which is expected as it is the most populous 
and economically active state in the country. This also 
explains why the West region leads in profitability (Q3).

## Overall Summary
This analysis of the Superstore Sales dataset reveals that 
Technology is the strongest performing category in both sales 
and profit margin. November is consistently the peak sales month 
driven by holiday shopping. The West region leads in profitability, 
largely driven by California which has the most orders nationally.

A key hidden insight is that while Office Supplies has a healthy 
overall profit margin, sub-categories like Appliances and Binders 
are severely loss-making and need urgent business attention. 
Furniture overall remains the weakest category with only 2.49% 
profit margin despite strong sales figures.
